### **Testing Hypothesis**

#### **~ Target Variable: House Sale Price**

1) Determine if living area affect housing price <br/>
2) Determine if there any correlation between overall quality of the house and price <br/>
3) Does Garage area influence the housing price? <br/>
4) Determine if there is correlation between the year house is built and cost <br/>
5) Is there any association between FullBath and SalePrice? <br/>


#### Checking the distribution of the variable of interest

```python
sns.distplot(house_price_dataset['SalePrice'], bins=50) #right skewed 
plt.savefig('../Results/sale_distribution.png', dpi=300, transparent=True)
plt.xticks(rotation=90)
```


#### Hypothesis 1: Determine if living area affect housing price


#### Using Lineplot to test hypothesis

```python
sns.set_style("darkgrid")
plt.figure(figsize=(8, 6))
sns.lineplot(x='GrLivArea', y='SalePrice', data=house_price_dataset)
plt.title('Line plot for relationship between Year Built - Sale Price')
plt.savefig('../Results/line_liv_area.png', dpi=300, transparent=True)
plt.xticks(rotation=90)
```

#### Using  scatter plot with a regression line to test hypothesis

```python
sns.lmplot(x='SalePrice', y='GrLivArea', data=house_price_dataset).set(title='Plot for relationship between Living Area - Sale Price ')
plt.xticks(rotation=90, fontsize=11)
plt.savefig('../Results/scatterplot_living_sale.png', dpi=300, transparent=True)
plt.xticks(rotation=90)
```

Conclusion:
Based on the scatter plot - regression line and lineplot, there is a strong positive correlation between living area and housing price. This suggests that as the living area increases, the price of the house tends to increase as well. Therefore, living area is a significant factor in determining the housing price.

#### Hypothesis 2: Determine if there any correlation between overall quality of the house and price


#### Using Regression plot to test hypothesis

```python
# set 3 hypothesis to test the house price
sns.regplot(x='OverallQual', y='SalePrice', data=house_price_dataset, robust=True).set(title='Plot for relationship between Overall Quality - Sale Price ')

plt.savefig('../Results/overall_sale_reg.png', dpi=300, transparent=True)
plt.xticks(rotation=90)
```

#### Using Boxplot to test hypothesis

```python
# Create box plot
sns.boxplot(x='OverallQual', y='SalePrice', data=house_price_dataset)
plt.title('Box plot for relationship between Overall Quality - Sale Price')
plt.savefig('../Results/boxplot_overall_sale.png', dpi=300, transparent=True)
plt.xticks(rotation=90)
```

Conclusion: Based on the regression plot and boxplot, there is a strong positive correlation between overall quality and price. This suggests that as the overall quality of the house increases, the price of the house tends to increase as well. Overall quality is a significant factor in determining the housing price.

#### Hypothesis 3: Does Garage area influence the housing price?


#### Using regression line and scatter plot to test hypothesis

```python
sns.lmplot(x='GarageArea', y='SalePrice', data=house_price_dataset).set(title='Plot for relationship between Garage - Sale Price ')

plt.savefig('../Results/garage_area_sale.png', dpi=300, transparent=True)
plt.xticks(rotation=90)
```

#### Using Line plot to test hypothesis

```python
sns.set_style("darkgrid")
plt.figure(figsize=(8, 6))
sns.lineplot(x='GarageArea', y='SalePrice', data=house_price_dataset)
plt.title('Line plot for relationship between Year Built - Sale Price')
plt.savefig('../Results/line_plot_year_sale.png', dpi=300, transparent=True)
plt.xticks(rotation=90)
```

Based on the scatterplot and lineplot, there is a positive correlation between garage area and housing price. This suggests that as the garage area increases, the price of the house tends to increase as well. However, there is a significant amount of variability in the data, and some houses with smaller garage areas also have high prices. Therefore, while garage area is a factor in determining housing price, it is not the only significant factor.

#### Hypothesis 4: Determine if there is correlation between the year house is built and cost


#### Using boxplot to test hypothesis

```python
# Create box plot
sns.set(rc={'figure.figsize':(25,20)})
sns.boxplot(x='YearBuilt', y='SalePrice', data=house_price_dataset)
plt.title('Box plot for relationship between Overall Quality - Sale Price', fontsize=20)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(rotation=90, fontsize=15)
plt.savefig('../Results/boxplot_year.png', dpi=300, transparent=True)
plt.xticks(rotation=90)
```

#### Using Jointplot to test hypothesis

```python
#  YearBuilt and SalePrice
jointplot_year = sns.jointplot(x='YearBuilt', y='SalePrice', data=house_price_dataset, kind='reg')
jointplot_year.fig.suptitle('Jointplot for YearBuilt - SalePrice', fontsize=15)

plt.savefig('../Results/jointplot.png', dpi=300, transparent=True)
plt.xticks(rotation=90)
```

The univariate plot of â€˜YearBuiltâ€™ shows that the distribution 
is skewed towards the year 2000 and has a long tail which extends till 1900.
The linear relationship between the variables is clearer in cases of recently built houses.

Based on the boxplot and joint plot-regression line, there is a weak positive correlation between the year the house is built and the housing price. This suggests that as the year the house was built increases, the price of the house tends to increase slightly. However, there is a lot of variability in the data, and some older houses also have high prices. Therefore, while the year the house is built is a factor in determining housing price, it is not a strong predictor.

#### Hypothesis 5: Is there any association between FullBath and SalePrice?

#### Using Jointplot to test hypothesis

```python
# FullBath and sale. negative correlation. no trend visible
jointplot_fullbath = sns.jointplot(x='SalePrice', y='FullBath', data=house_price_dataset)
jointplot_fullbath.fig.suptitle('Jointplot for FullBath - SalePrice', fontsize=15)
plt.savefig('../Results/jointplot_fullbath_sale.png', dpi=300, transparent=True)
plt.xticks(rotation=90)
```

#### Using boxplot plot to test hypothesis

```python
# Create box plot
sns.set(rc={'figure.figsize':(7, 8)})
sns.boxplot(x='FullBath', y='SalePrice', data=house_price_dataset)
plt.title('Box plot for relationship between FullBath - Sale Price', fontsize=20)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(rotation=90, fontsize=15)

plt.savefig('../Results/boxplot_fullbath.png', dpi=300, transparent=True)
plt.xticks(rotation=90)
```

Based on the boxplot and jointplot, there is a weak positive correlation between FullBath and SalePrice. This suggests that as the number of FullBaths increases, the price of the house tends to increase slightly. However, there is a lot of variability in the data, and some houses with fewer FullBaths also have high prices. Therefore, while FullBath is a factor in determining housing price, it is not a strong predictor.

### Using Correlation matrix and heatmap to test hypothesis

```python
# Create heatmap
sns.set(font_scale=1.4)
sns.heatmap(house_price_dataset[['OverallQual', 'SalePrice', 'GrLivArea', 'YearBuilt', 'FullBath']].corr(), annot=True)
plt.title('Heatmap for correlation between Overall Quality, Sale Price, GrLivArea, YearBuilt')
plt.savefig('../Results/heatmap_for_housing.png', dpi=300, transparent=True)
```

### Conclusion of Hypothesis

Overall, analysis of the housing price dataset reveals that several factors impact the price of a house. Among these, overall quality, living area and garage area are the strongest predictors of housing price. A larger living area tends to correlate with a higher sale price, while houses with better quality also generally have higher prices. On the other hand, the impact of year the house was built, and FullBath on housing prices is comparatively weaker. While houses with larger garages or more FullBath tend to have slightly higher prices, the relationship is not as strong as that of living area or overall quality. 

From the 5 hypothesis stated, living area and overall quality are strong predictors of housing price, while garage area, the year the house was built, and FullBath are weaker predictors.
